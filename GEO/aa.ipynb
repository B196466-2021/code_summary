```python

file_extension = ".xlsx"
txt_extension = "_series_matrix.txt"

# 定义要提取的行标题和起始/结束行的行名
start_row_title = "!Sample_geo_accession"
end_row_title = "!Sample_platform_id"

# 初始化计数器变量和起始/结束行的行号
i = 0
start_row_number = None
end_row_number = None


# 检测文件编码格式
#with open(GEO_accession+txt_extension, 'rb') as f:
#    result = chardet.detect(f.read())
#    encoding = result['encoding']

# 打开文件并逐行读取内容
with open(GEO_accession+txt_extension, "r") as f:
    for line in f:
        # 去掉行尾的换行符
        line = line.rstrip("\n")
        
        # 如果当前行包含起始行标题，记录起始行号
        if start_row_title in line:
            start_row_number = i
        
        # 如果当前行包含结束行标题，记录结束行号并退出循环
        if end_row_title in line:
            end_row_number = i
            break
        
        # 更新计数器变量
        i += 1

# 如果找到了起始行和结束行的行号，则读取文件内容，并提取指定的行
if start_row_number is not None and end_row_number is not None:
    with open(GEO_accession+txt_extension, "r") as f:
        # 跳过起始行之前的行
        for i in range(start_row_number):
            next(f)
        
        # 读取需要的行并保存到 DataFrame 中
        data = []
        for i in range(start_row_number, end_row_number+1):
            line = next(f).rstrip("\n")
            data.append(line.split("\t"))
        df = pd.DataFrame(data)
        
        # 将 DataFrame 转置并保存到 Excel 文件中
        df_transposed = df.T
        df_transposed.to_excel(GEO_accession + file_extension, index=False, header=False)
```

```python
import openpyxl

# 打开 Excel 文件
wb = openpyxl.load_workbook(GEO_accession + file_extension)
sheet = wb.active

# 定义要删除的列名列表
cols_to_delete_names = ["!Sample_status", "!Sample_submission_date", "!Sample_last_update_date", "!Sample_type", "!Sample_channel_count", "!Sample_source_name_ch1", "!Sample_organism_ch1", "!Sample_taxid_ch1"]

# 遍历每一列，检查是否为要删除的列
cols_to_delete = []
for col in sheet.columns:
    for col_name in cols_to_delete_names:
        if col[0].value == col_name:
            cols_to_delete.append(col[0].column)

# 倒序删除要删除的列
cols_to_delete.sort(reverse=True)
for col_idx in cols_to_delete:
    sheet.delete_cols(col_idx)

# 保存修改后的 Excel 文件
wb.save(GEO_accession + file_extension)

```

```python
from openpyxl import load_workbook
from openpyxl.utils.cell import get_column_letter

# 打开 Excel 文件
workbook = load_workbook(filename=GEO_accession + file_extension)

# 获取要处理的工作表
worksheet = workbook.active

# 创建一个新的工作表对象
new_worksheet = workbook.create_sheet(title='Processed Data')

# 获取工作表的最大行和列数
max_row = worksheet.max_row
max_column = worksheet.max_column

# 遍历工作表中的所有单元格
for row in range(1, max_row+1):
    for col in range(1, max_column+1):
        # 获取单元格的字母表示
        col_letter = get_column_letter(col)
        # 获取单元格的值
        cell_value = worksheet.cell(row=row, column=col).value
        # 使用 replace() 方法去除双引号
        if isinstance(cell_value, str):
            cell_value = cell_value.replace('"', '')
        # 将去除双引号后的值写入新的工作表对象
        new_worksheet.cell(row=row, column=col, value=cell_value)

# 删除原始工作表对象
workbook.remove(worksheet)

# 保存 Excel 文件
workbook.save(filename=GEO_accession + file_extension)
```

```python
##Single
def clinic_info_process(dt):
    f=pd.read_excel(dt,engine='openpyxl').fillna(value='NA')
    #去掉列名中的'!Sample_'
    for x in f.columns:
        l=re.sub('!Sample_','',x)
        f=f.rename(columns={x:l})
    #提取行名中：前面的字段并修改列名为该字段
    for x in f.columns:
        if 'characteristics_ch1' in x:
            l=f[x].iloc[0].split(': ')[0]
            f=f.rename(columns={x:l})
    # 去掉列名中的'_ch1'
    for x in f.columns:
        if '_ch1' in x:
            l=re.sub('_ch1','',x)
            f=f.rename(columns={x:l})
    #清理每一行的字符
    for x in f.columns:
        if ': ' in f[x].iloc[0]:
            f[x]=f[x].str.replace(x+': ','',regex=False)
    #save
    f.to_excel(dt.split('.')[0]+'_info.xlsx',sheet_name=dt.split('_')[0],index=None)
```
